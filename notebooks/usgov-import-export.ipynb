# Learn about world trade by looking at US Census import / export data

## Why this data is interesting

The US government keeps very detailed records on all goods being import or exported from the United States.

Want to estimate the size of a world market or see if it's growing or shrinking?

Want to find the best place in the world to manufacture a particular product?.. or get a sense of how much importers are paying for the goods they wholesale? This dataset can give you a lot of valuable and interesting insights into how the world economy works.

It's also super easy to access and it's government data. It's pretty darn clean!

## Ideas for questions this data can help you answer

* Where is the best place in the world to have a particular type of product manufactured?
* What kinds of products are Americans buying?
* Where are they importing goods from?
* What do they sell and where do they sell to?
* How much do products cost before they get marked up?

## Where you can go to get this kind of data

This data is available through the US census beaureau's website.

https://usatrade.census.gov/index.php

They also provide access to it via API!!!

https://www.census.gov/developers/

The data is super easy to work with. Definitely worth checking out.
