## Why this data is interesting

Ever wondered how Google Analytics works? It logs events - visits, clicks, sigunps, etc... and then aggregates them together and plots them.

When companies grow (or when they have complicated reporting requirements) they often out-grow solutions like Google Analytics and end up having to roll their own solutions.

If you work as a data analyst in a big tech company chances are you'll spend a lot of time writing queries that grab events from a database... group them into bins (maybe by month or something) and then plot them.

If you're interested in learning more about what enterprises are using these days I recommend checking out [Looker](https://looker.com/)

This is a bit of log data from a poetry community app that I run.

Try grouping events over time and plotting them.


## Ideas for questions this data can help you answer

* How many poems are being created per day?
* How many poems are being deleted per day?
* How are the stats trending over time?
* Am I making any money (answer: no).
